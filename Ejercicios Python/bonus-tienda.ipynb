Crea un programa nuevo que permita al usuario ingresar varios art√≠culos, cada uno con su precio original y cup√≥n de descuento. Aplica los descuentos correspondientes a cada producto y muestra el monto total final de la compra.

```python
def ingresar_articulo():
    try:
        precio_original = 0
        while precio_original <= 0:
            precio_original = float(input("Introduce el precio original: "))

            if precio_original < 0:
                print("El precio ha de ser un valor positivo")

    except ValueError:
        print("Introduce un n√∫mero v√°lido")

    try:
        descuento = float(input("Introduce el valor del descuento en %: "))

        if descuento < 0 or descuento > 100:
            print("El descuento ha de estar entre 0 y 100")

        elif descuento < 5:
            descuento = 5
            print("El descuento no puede ser menor que 5%, se le aplicar√° un 5%")

        elif descuento > 50:
            descuento = 50
            print("No te me flipes, te pongo m√°ximo un 50%")

    except ValueError:
        print("Introduce un n√∫mero v√°lido")

    precio = precio_original * (100 - descuento) / 100
    print("El precio de este art√≠culo es: ", precio)
    return precio

try:
    n = 0
    while n < 1:
        n = int(input("¬øCu√°ntos art√≠culos vas a meter? "))

        if n < 1:
            print("Debes introducir al menos un art√≠culo")

except ValueError:
    print("Tienes que introducir un n√∫mero entero")

precio = 0
for i in range(n):
    print("Ingresa el producto n√∫mero: ", i + 1)
    precio += ingresar_articulo()

print("El valor final de la cesta es: ", precio)
```
